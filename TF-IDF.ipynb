TF/IDF is short for Term-Frequency/Inverse-Document-Frequency. It's the score generated by taking the ratio of the frequency of a term in document $a$ over that word's frequency in the entire corpus. It's a tool for information extraction--generally to find topical information--but we'll use it to find defining aspects of content.

```python
# these are the imports we'll be using
import glob
import matplotlib.pyplot as plt
import numpy as np
import book
import pandas
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
```

```python
postmodern = ''
post911 = ''
tf = TfidfVectorizer(analyzer='word', ngram_range=(1,3), min_df = 0, stop_words = 'english')

for filename in glob.glob('postmodern/*.txt'):
    try:
        postmodern += book.makebook(open(filename, 'r')).text
    except Exception as e:
        print(book)

for filename in glob.glob('post911/*.txt'):
    try:
        post911 += book.makebook(open(filename, 'r')).text
    except Exception as e:
        print(book)

corpus = [postmodern, post911]
```

```python
tfidf_matrix =  tf.fit_transform(corpus)
tfidf_matrix =  tf.fit_transform(corpus)
feature_names = tf.get_feature_names() 
```

```python
dense = tfidf_matrix.todense()
pmod = dense[0].tolist()[0]
pnin = dense[1].tolist()[0]
phrase_scores_pmod = [pair for pair in zip(range(0, len(pmod)), pmod) if pair[1] > 0]
phrase_scores_pnin = [pair for pair in zip(range(0, len(pnin)), pnin) if pair[1] > 0]
```

```python
sorted_phrase_scores_pmod = sorted(phrase_scores_pmod, key=lambda t: t[1] * -1)
for phrase, score in [(feature_names[word_id], score) for (word_id, score) in sorted_phrase_scores_pmod][:20]:
   print('{0: <20} {1}'.format(phrase, score))
```

```python
sorted_phrase_scores_pnin = sorted(phrase_scores_pnin, key=lambda t: t[1] * -1)
for phrase, score in [(feature_names[word_id], score) for (word_id, score) in sorted_phrase_scores_pnin][:20]:
   print('{0: <20} {1}'.format(phrase, score))
```
